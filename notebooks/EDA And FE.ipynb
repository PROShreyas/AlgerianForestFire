## About Dataset
#### The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

#### 122 instances for each region.

#### The period from June 2012 to September 2012.
#### The dataset includes 11 attributes and 1 output attribute (class). The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on the region 
1. Bejaia Region dataset
2. Sidi-Bel Abbes Region Dataset
##### Add new column with region

```python
df = dataset
df.loc[:122,"Region"]=0
df.loc[122:,"Region"] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## remove the null values 
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in the column names
df.columns = df.columns.str.strip()
df.columns
```

#### Change the required column as integer data type 

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
df.info()
```

#### Changing the other columns to float datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_Forest_Cleaned_Dataset.csv',index=False)
```
